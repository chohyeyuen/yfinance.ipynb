```python
import requests
import json
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
#getting url
URL = "https://topforeignstocks.com/indices/the-components-of-the-korea-stock-exchange-kospi-index/"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
print (soup.prettify())
```

```python
#finding table
table = soup.find("table")
table
```

```python
print(table.findAll('tr'))
```

```python
#making columns and assigning values to the corresponding colunm
S_NO=[]
Name=[]
Code=[]
rows=table.findAll('tr')

for row in rows:
    cells=row.findAll('td')
    if(len(cells)==3):
        S_NO.append(cells[0].find(text=True))
        Name.append(cells[1].find(text=True))
        Code.append(cells[2].find(text=True))
```

```python
Name
```

```python
#creating df1
df1 = pd.DataFrame()
df1['S_NO']=S_NO
df1['Name']=Name
df1['Code']=Code

df1
```

```python
#adding ".KS" to the stock code since thats how yfinance read Korean stock
df1['Code']=df1['Code']+'.KS'
```
